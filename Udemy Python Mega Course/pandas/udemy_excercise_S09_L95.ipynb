```python
# extend data frame with a new column 'Region' and populate with 'North America' value- Note: this is an in-place operation
# new column data must be inserted into every row -> df5.shape[0] - gives number of rows in df5
df5=pandas.read_csv('supermarkets.csv')
df5['Region']=df5.shape[0]*["North America"]
df5
```

```python
# Modify Region column by including value of another column in the data frame
df5['Region']=df5['Country']+' - '+'Canada'
df5
```

```python
# Merge columns into a new column i.e. 'Postal Address'
df6 = df5
df6['Postal address']=df6['Address']+', '+df6['City']+', '+df6['State']+', '+df6['Country']
df6
```

```python
# transpose data frame
df5_t = df5.T
df5_t
```

```python
# add new row to data frame -> transpose df, add new column, transpose back
df7=df5.T
df7[6]=['7','1 High St','Ennis','Clare','Ireland','Dunnes Store','15','Europe','']
df7 = df7.T

# regenerate 'Postal address' column
df7['Postal address'] = df7['Address']+', '+df7['City']+', '+df7['State']+', '+df7['Country']
df7
```

```python
# fetch coordinates based on postal address

# pip install geopy
from geopy.geocoders import Nominatim
nom=Nominatim(scheme='http')
df7['Coordinates']=df7['Postal address'].apply(nom.geocode)
df7['Latitude']=df7['Coordinates'].apply(lambda x: x.latitude if x != None else None)
df7['Longitude']=df7['Coordinates'].apply(lambda x: x.longitude if x != None else None)
df7
```
